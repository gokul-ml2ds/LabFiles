```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset from : https://archive.ics.uci.edu/ml/datasets/student+performance
data = pd.read_csv('student-por.csv',sep=';')
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
# Dropping NaN values to clean the dataset
data.dropna(axis=0,inplace=True)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['Mjob']= label_encoder.fit_transform(data['Mjob'])
data['Fjob']= label_encoder.fit_transform(data['Fjob'])
data['reason']= label_encoder.fit_transform(data['reason'])
data['guardian']= label_encoder.fit_transform(data['guardian'])
```

```python
data.describe()
```

```python
sns.countplot(data.sex)
```

```python
data.info()
```

```python
# Check if there are any missing data points
# data.isnull().sum()
data.isnull().values.any()
```

```python
# Plotting final scores
plt.style.use('ggplot')
data['G3'].plot.hist(title='Histogram of G3 Grades', bins=20)
plt.xlabel('Grades - G3')
```

```python
# Boxplot to check for outliers
sns.boxplot(x=data.G3)
```

```python
# Calculating the Z score
from scipy import stats 
z = np.abs(stats.zscore(data['G3']))
np.where(z>3)
```

```python
# Finding and Removing outlier data
np.where(data['G3']<2)
```

```python
# From the box plot and z-score, we get the indexes of outlier data points
data.drop(np.where(z>3)[0],inplace=True)
```

```python
sns.countplot(data.age)
```

```python
# Relation between studytime and grade
plt.scatter(data.studytime,data.G3)
plt.xlabel('Number of hours studied')
plt.ylabel('Final grade')
plt.show()
```

```python
data["school"].value_counts()
```

```python
# Correlation between study time and final grade
data.corr()['studytime']['G3']
```

```python
# Correlation between absencce and final grade
data.corr()['absences']['G3']
```
