```python
# importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_datareader import wb, data
import datetime
```

```python
# downloading data for apple stocks from yahoo finance from 2015 to 2020
ticker = 'AAPL'
start_date = '2015-01-01'
end_date = '2020-01-01'

# User pandas_reader.data.DataReader to load the desired data.
panel_data = data.DataReader(ticker,'yahoo',start_date, end_date)
```

```python
# Prining the data
panel_data = pd.DataFrame(panel_data)
panel_data.head()
```

```python
# data sources
panel_data.columns
```

```python
# summarize the data
panel_data.describe()
```

```python
plt.figure(figsize=(16,8))
plt.plot(panel_data.High)
plt.xlabel('Date')
plt.ylabel('stock price in dollars')
plt.title('Apple Stock Prices')
```

```python
plt.plot(panel_data.Volume)
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Apple Stock Volume')
```
